##Linear Regresion theory

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
#Declaramos los vectores
X=[173,171,189,181] #altura
y=[81,72,96,94] #peso

V_inicial = np.c_[np.ones(4), X]

#aplicamos la ec normal
theta = np.linalg.inv(V_inicial.T.dot(V_inicial)).dot(V_inicial.T).dot(y)
theta
```

```python
plt.scatter(X,y,s=40, c="blue")
x_lim = [170,190]
x_lim_unos= np.c_[np.ones(2), x_lim]
y_lim = x_lim_unos.dot(theta)
plt.plot(x_lim, y_lim)
```

```python
#predecimos
altura = 175
peso = theta[0] + (theta[1]*altura)
peso
```

```python
#Utilizando sklear

regresion = linear_model.LinearRegression()

altura_sklear = np.array(X).reshape((-1,1))

modelo = regresion.fit(altura_sklear, y)

print('theta_0', modelo.intercept_) #interseccion con el eje de las y
print('theta_1', modelo.coef_) #pendiente

entrada = [[120], [190], [161], [174]]

modelo.predict(entrada)

plt.scatter(entrada,modelo.predict(entrada), s=40, c="blue")
plt.plot(entrada, modelo.predict(entrada))
plt.ylabel("peso")
plt.xlabel("altura predictiva")
plt.scatter(X,y, color = "pink")
```
